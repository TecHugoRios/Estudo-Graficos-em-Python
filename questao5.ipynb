Você é analista de dados de uma plataforma de e-commerce e foi solicitado a analisar as vendas de um produto específico ao longo do ano. Os dados de vendas mensais para esse produto são:

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dados = {'Mês':["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", "Outubro", "Novembro", "Dezembro"],
         'Vendas (Unidades)':[150,200, 180, 220, 250, 210, 270, 300, 350, 400, 450, 500]}
```

```python
df = pd.DataFrame(dados)
df
```

```python
plt.figure(figsize=(14,8))
plt.plot(df['Mês'], df['Vendas (Unidades)'], color='red', linestyle=':',linewidth=3 ,label='Variação', marker=".", markersize=15)

plt.ylabel("Vendas (Unidades)")
plt.xlabel("Mês")
plt.show()
```
