Como parte de uma análise de preços de produtos em um e-commerce, você recebeu um conjunto de dados contendo os preços de vários itens. Os preços dos produtos são:

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dados = {
    'Dia': ["01/01/24", "02/01/24", "03/01/24", "04/01/24", "05/01/24", "06/01/24", "07/01/24", "08/01/24"],
    'Preço (R$)': [30.00, 31.00, 29.50, 32.00, 31.50, 33.00, 34.00, 33.50],
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
calcVariacao = [dAtual - dAnterior for dAnterior, dAtual in zip(df['Preço (R$)'][:-1], df['Preço (R$)'][1:])]
calcVariacao
```

```python
plt.figure(figsize=(14,8))
plt.plot(df['Dia'], df['Preço (R$)'], color='red', linestyle=':',linewidth=3 ,label='Variação', marker=".", markersize=15)

for item, (x,y) in enumerate(zip(df['Dia'][1:], df['Preço (R$)'][1:])):
    plt.annotate(format(((calcVariacao[item]/df['Preço (R$)'].sum()) * 100), ".1f")+ '%', xy=(x, y), xytext=(0, 10), textcoords='offset points', color='black')

plt.ylabel("Preço (R$)")
plt.xlabel("Variação Diária")
plt.show()
```
