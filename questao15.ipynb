Você trabalha em uma empresa de varejo e precisa prever a demanda de produtos para a próxima semana. Você tem acesso a dados históricos de vendas semanais. Os dados são:
  Calcule a média de vendas semanais e crie uma visualização que mostre essa demanda ao longo das semanas. Utilize essas informações para prever a quantidade de produtos necessária para manter o estoque abastecido na próxima semana.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dados = {
    'Semana': [1, 2, 3, 4],
    'Vendas (Unidades)': [200, 220, 250, 270]
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
mediaVendas = float(df['Vendas (Unidades)'].mean())
mediaVendas
```

```python
plt.plot(df['Semana'],df['Vendas (Unidades)'], marker='o')
plt.axhline(mediaVendas, color='red',linestyle='--')
plt.title('Vendas Semanais e Média de Vendas')
plt.legend
plt.show()
```
