Você está analisando os dados de um centro de atendimento ao cliente e precisa entender o tempo médio necessário para resolver os chamados. Os tempos de resolução dos chamados são:


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dados = {
    'Chamado ID': ["001", "002", "003", "004", "005", "006"],
    'Tempo de Resolução (h)': [2.5, 3.0, 1.0, 4.0, 1.5, 5.0],
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
tempoMedio = df['Tempo de Resolução (h)'].mean()
tempoMedio = float(format(tempoMedio, ".2f"))
tempoMedio
```

```python
plt.figure(figsize=(14,8))
plt.plot(df['Chamado ID'], df['Tempo de Resolução (h)'], color='red', linestyle=':',linewidth=3 ,label='Variação', marker=".", markersize=15)

for i, (x, y) in enumerate(zip(df['Chamado ID'], df['Tempo de Resolução (h)'])):
    plt.annotate(f"{y}h", xy=(x, y), xytext=(-10, 10), textcoords='offset points', color='black')

plt.show()
```
