Você foi convidado a realizar uma análise dos funcionários de uma empresa com o objetivo de entender a relação entre idade e salário. O conjunto de dados contém informações sobre a idade e o salário de cada funcionário:

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dados = {'Funcionário':["1", "2", "3", "4", "5", "6", "7", "8"],
         'Idade':[25,32, 28, 40 , 50, 35, 29, 45],
         'Salário (R$)':[3000, 4500, 3800, 7000,10000,5200,4000,8500]
         }
```

```python
df = pd.DataFrame(dados)
df
```

```python

df.plot.scatter(x='Funcionário',y='Idade', c='Salário (R$)', colormap='viridis')
plt.show()
```
